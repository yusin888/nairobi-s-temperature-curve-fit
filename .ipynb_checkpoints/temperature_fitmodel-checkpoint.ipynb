```python
x= date2; y=average_temperature
#y= date2; x=average_temperature
```

```python
def line(x, a, b, c):
    return (a*np.cos(x*(2*3.142*x + b))+ c)
```

```python
popt, pcov=curve_fit(line, x, y) 
```

```python
a, b, c= popt
```
